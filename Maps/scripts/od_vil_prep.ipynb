```python
import geopandas as gpd
import pandas as pd
import numpy as np
import regex as re
```

```python
od_block = gpd.read_file(r'D:\CivicDataLab_IDS-DRR\IDS-DRR_Github\flood-data-ecosystem-Odisha\Maps\od_ids-drr_shapefiles\odisha_block_final.geojson')
od_block
```

```python
od_vil_unfiltered = gpd.read_file(r'D:\CivicDataLab_IDS-DRR\IDS-DRR_Github\flood-data-ecosystem-Odisha\Maps\od_ids-drr_shapefiles\od_villages.geojson')
od_vil_unfiltered
```

```python
od_vil = od_vil_unfiltered.replace(r'^\s*$', np.nan, regex=True)
od_vil = od_vil.dropna(subset=['vilnam_soi'])
od_vil = od_vil.dropna(subset=['vilname11'])
od_vil
```

```python
od_vil = od_vil.loc[od_vil["stname"] == "ODISHA"]
od_vil["vilnam_soi"] = od_vil["vilnam_soi"].str.upper()
od_vil = od_vil[~od_vil["vilnam_soi"].str.contains("FOREST", na=False)]
od_vil = od_vil[~od_vil["vilnam_soi"].str.contains("R.F", na=False)]
od_vil = od_vil[~od_vil["vilnam_soi"].str.contains("D.P.F", na=False)]
od_vil = od_vil[~od_vil["vilnam_soi"].str.contains("JUNGLE", na=False)]
od_vil = od_vil[~od_vil["vilnam_soi"].str.contains(" HILL", na=False)]
od_vil = od_vil[~od_vil["vilnam_soi"].str.contains("R F", na=False)]
od_vil
```

```python
od_vil['dtcode11'] = od_vil['dtcode11'].astype(str).apply(lambda x: x.zfill(2))
od_vil['block_lgd'] = od_vil['block_lgd'].astype(float).astype(int).astype(str).apply(lambda x: x.zfill(5))
od_vil['object_id'] = od_vil.apply(lambda row: f"21-{row['dtcode11']}-{row['block_lgd']}", axis=1)
od_vil
```

```python
od_vil.columns
```

```python
od_vil = od_vil.drop(['objectid','remark', 'ac_no', 'test' ],axis=1)
od_vil.to_file(r'D:\CivicDataLab_IDS-DRR\IDS-DRR_Github\flood-data-ecosystem-Odisha\Maps\od_ids-drr_shapefiles\od_villages_final.geojson')
```

### Urban shapes

```python
od_urban = gpd.read_file(r'D:\CivicDataLab_IDS-DRR\IDS-DRR_Github\flood-data-ecosystem-Odisha\Maps\Geojson\od_urban_shapes.geojson')
od_urban
```

```python
od_urban = od_urban.dropna(subset=['VILNAM_SOI'])
od_urban = od_urban.loc[od_urban["STCODE11"] == "21"]
od_urban["VILNAM_SOI"] = od_urban["VILNAM_SOI"].str.upper()
od_urban = od_urban[~od_urban["VILNAM_SOI"].str.contains("FOREST", na=False)]
od_urban = od_urban[~od_urban["VILNAM_SOI"].str.contains("R.F", na=False)]
od_urban = od_urban[~od_urban["VILNAM_SOI"].str.contains("D.P.F", na=False)]
od_urban = od_urban[~od_urban["VILNAM_SOI"].str.contains("JUNGLE", na=False)]
od_urban = od_urban[~od_urban["VILNAM_SOI"].str.contains(" HILL", na=False)]
od_urban = od_urban[~od_urban["VILNAM_SOI"].str.contains("R F", na=False)]

od_urban
```

```python
# https://en.wikipedia.org/wiki/List_of_cities_in_Odisha_by_population
urban_filter = ['Bhubaneswar','Cuttack','Rourkela','Berhampur','Sambalpur','Puri','Balasore','Bhadrak','Baripada','Bargarh','NAC','CT','IT','OG']
filtered_gdf = od_urban[od_urban['VILNAME11'].str.contains('|'.join(urban_filter), na=False)]
filtered_gdf
```

```python
merged = filtered_gdf.merge(od_vil[['object_id','vilcode11']],left_on='VILCODE11', right_on='vilcode11',how='left')
merged
```

```python
merged = merged.drop('vilcode11',axis=1)
final_gdf = gpd.GeoDataFrame(merged,geometry='geometry')
final_gdf.to_file(r'D:\CivicDataLab_IDS-DRR\IDS-DRR_Github\flood-data-ecosystem-Odisha\Maps\od_ids-drr_shapefiles\od_urban_final.geojson')
```
